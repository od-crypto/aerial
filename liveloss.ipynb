```python
import random
import time
plotlosses = PlotLosses()
for i in range(10):
    if i % 5 == 0:
        plotlosses.update({'val_acc': 0.4, 'loss': 0.9})
    else:
        plotlosses.update({'acc': random.random(), 'loss': 0.9})
    plotlosses.send()
    time.sleep(1)
```
